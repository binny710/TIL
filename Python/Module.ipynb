### **모듈 임포트**
 + 그동안 사용했던 함수들 처럼, 다양한 기능들이 미리 함수로 구현되어 모듈 형태로 제공
 + 대표적으로 추후 과정에서 사용하게 될 아래의 모듈들이 존재

 + requests - HTTP 요청/응답 모듈
 + numpy - 수치해석 모듈 
 + pandas - 데이터 분석 모듈


#### import 
 - import를 사용하여 해당 모듈 전체를 import

```python
import requests
resp = requests.get('http://naver.com')
resp.text  #html
```

```python
import math
math.pi
```

```python
math.cos(2*math.pi)
```

#### from import 
 - 해당 모듈에서 특정한 타입만 import

```python
from math import sin
```

```python
sin(2)       #앞에 math를 붙일필요 없다. 
```

#### \* 임포트
 - 해당 모듈내에 정의된 모든 것을 import
 - 일반적으로 사용이 권장되지 않음: 이름이 중복되는 경우가 있기 때문

```python
cos(1)
```

```python
from math import *
```

```python
cos(1)
```

#### as 
 - 모듈 import 시, alias(별명) 지정가능
 

```python
import pandas as pd
```

```python
import numpy as np
```
